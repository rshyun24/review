```python
# 1. 데이터 로드

import pandas as pd

df_pc     = pd.read_csv(r"C:\lecture\flow\00_data\00_raw\paulaschoice_ingredients.csv", encoding="utf-8")
df_coos   = pd.read_excel(r"C:\lecture\flow\00_data\00_raw\coos_성분정보.xlsx")
df_hwahae = pd.read_csv(r"C:\lecture\flow\00_data\00_raw\hwahae_all.csv", encoding="utf-8")

print("=== PaulasChoice 원본 컬럼 ===")
print(df_pc.columns.tolist())
print(f"Shape: {df_pc.shape}\n")

print("=== COOS 원본 컬럼 ===")
print(df_coos.columns.tolist())
print(f"Shape: {df_coos.shape}\n")

print("=== Hwahae 원본 컬럼 ===")
print(df_hwahae.columns.tolist())
print(f"Shape: {df_hwahae.shape}\n")
```

```python
# 2. PaulasChoice — 컬럼 드롭
pc_drop_cols = ["연관성분", "참고논문"]

df_pc = df_pc.drop(columns=pc_drop_cols)

print("=== PaulasChoice 정리 후 컬럼 ===")
print(df_pc.columns.tolist())
print(f"Shape: {df_pc.shape}\n")
```

```python
# 3. COOS — 컬럼 드롭
coos_drop_cols = [
    "URL",                  # 분석 불필요
    "구명칭",                # 대부분 '-', 이명과 중복
    "EC No.",               # CAS No.와 중복, 활용도 낮음
    "구조식",                # 대부분 '-' 또는 결측, 피처화 불가
    "유럽 CosIng Ref No.",   # 분석 범위 외
    "링크",                  # URL과 동일 역할
    "한글명",
    "이명"
]

df_coos = df_coos.drop(columns=coos_drop_cols)

print("=== COOS 정리 후 컬럼 ===")
print(df_coos.columns.tolist())
print(f"Shape: {df_coos.shape}\n")
```

```python
# 4. Hwahae — 컬럼 드롭
hwahae_drop_cols = [
    "discount_rate",
    "capacity",
    "sub_attrs",
    "sub_product_name",
    "concentration_info",
    "skin_type",
    "skin_remark_good",
    "skin_remark_bad"
]

df_hwahae = df_hwahae.drop(columns=hwahae_drop_cols)

print("=== Hwahae 정리 후 컬럼 ===")
print(df_hwahae.columns.tolist())
print(f"Shape: {df_hwahae.shape}\n")
```

```python
# PaulasChoice 결측치 제거
df_pc.info()
df_pc = df_pc.dropna(axis=0, how='any')
```

```python
# coos '스코어', '데이터 등급' -> 0으로 채우기
df_coos['스코어'] = df_coos['스코어'].fillna('0')
df_coos['데이터 등급'] = df_coos['데이터 등급'].fillna('0')
df_coos.info()
```

```python
# coos 결측치 제거
df_coos = df_coos.dropna(axis=0, how='any')
df_coos.info()
```

```python
# Hwahae 'primary_attr'의 결측치를 '클렌징'으로 채우기
df_hwahae['primary_attr'] = df_hwahae['primary_attr'].fillna('클렌징')
```

```python
# Hwahae 결측치 제거
df_hwahae = df_hwahae.dropna(axis=0, how='any')
df_hwahae.info()
```

```python
# 한글 성분명 → 'ingredient_ko'로 통일
df_pc = df_pc.rename(columns={"한글명": "ingredient_ko"})
df_coos = df_coos.rename(columns={"성분명": "ingredient_ko"})
df_hwahae = df_hwahae.rename(columns={"korean": "ingredient_ko"})

# 영문 성분명 → 'ingredient_en'로 통일
# ※ 각 파일의 실제 영문 컬럼명에 맞춰 아래 키를 수정하세요
df_pc = df_pc.rename(columns={"영문명": "ingredient_en"})           # PaulasChoice
df_coos = df_coos.rename(columns={"INCI": "ingredient_en"})        # COOS
df_hwahae = df_hwahae.rename(columns={"english": "ingredient_en"}) # Hwahae

# 공백 제거 (병합 시 매칭 누락 방지)
for df in [df_pc, df_coos, df_hwahae]:
    df["ingredient_ko"] = df["ingredient_ko"].astype(str).str.strip()
    df["ingredient_en"] = df["ingredient_en"].astype(str).str.strip()
```

```python
# 공통 키 2개(ingredient_ko, ingredient_en)는 prefix 제외
key_cols = ["ingredient_ko", "ingredient_en"]

df_pc = df_pc.rename(columns=lambda c: f"pc_{c}"   if c not in key_cols else c)
df_coos = df_coos.rename(columns=lambda c: f"coos_{c}" if c not in key_cols else c)
df_hwahae = df_hwahae.rename(columns=lambda c: f"hw_{c}" if c not in key_cols else c)

# 한글명 + 영문명 둘 다를 키로 사용해 outer join
df_merged = (
    df_pc
    .merge(df_coos,   on=key_cols, how="outer")
    .merge(df_hwahae, on=key_cols, how="outer")
)

print(f"병합된 shape: {df_merged.shape}")
print("현재 컬럼 목록:")
print(df_merged.columns.tolist())
```

```python
rename_to_english = {
    # --- PaulasChoice ---
    "pc_등급" : "pc_rating",            # 등급
    "pc_효과별" : "pc_effect",          # 효과별 분류
    "pc_분류" : "pc_category",          # 분류
    "pc_성분설명" : "pc_description",    # 성분 설명

    # --- COOS ---
    "coos_기능" : "coos_function",              # 기능
    "coos_종류" : "coos_type",                  # 종류
    "coos_CAS No." : "coos_cas_no",            # CAS 번호
    "coos_국가" : "coos_country",               # 국가
    "coos_🇰🇷국내" : "coos_kr_restricted",       # 국내(한국) 규제 여부
    "coos_🇨🇳중국" : "coos_cn_restricted",       # 중국 규제 여부
    "coos_🇹🇼대만" : "coos_tw_restricted",       # 대만 규제 여부
    "coos_🇯🇵일본" : "coos_jp_restricted",        # 일본 규제 여부
    "coos_🇩🇪유럽" : "coos_eu_restricted",        # 유럽 규제 여부
    "coos_🇻🇳아세안" : "coos_asean_restricted",   # 아세안 규제 여부
    "coos_AI설명" : "coos_ai_description",      # AI 설명
    "coos_스코어" : "coos_score",                # 스코어
    "coos_데이터 등급" : "coos_data_grade",      # 데이터 등급
}

df_merged = df_merged.rename(columns=rename_to_english)
print(df_merged.columns.tolist())
```

```python
drop_cols = ["hw_ingredient_id", "hw_is_allergy", "hw_purposes", "coos_cas_no", "coos_country"]

df_merged = df_merged.drop(columns=drop_cols, errors='ignore')

print(f"최종 shape: {df_merged.shape}")
print(df_merged.columns.tolist())
```

```python
import numpy as np

# ── coos_score 변환 ──────────────────────────────────────
def map_coos_score(val):
    if pd.isna(val) or str(val).strip() == "":
        return 0
    val = str(val)
    if "안전" in val:
        return 1
    elif "주의" in val:
        return 2
    elif "위험" in val:
        return 3
    else:
        return 0

# ── pc_rating 변환 ───────────────────────────────────────
def map_pc_rating(val):
    if pd.isna(val) or str(val).strip() == "":
        return 0
    val = str(val).strip()
    mapping = {
        "훌륭함": 1,
        "좋음":   2,
        "보통":   3,
        "나쁨":   4,
        "매우 나쁨": 5,
    }
    return mapping.get(val, 0)  # 매핑 안되면 0

df_merged["coos_score"]  = df_merged["coos_score"].apply(map_coos_score)
df_merged["pc_rating"]   = df_merged["pc_rating"].apply(map_pc_rating)

# 확인
print(df_merged["coos_score"].value_counts())
print(df_merged["pc_rating"].value_counts())
```

```python
import json
import numpy as np

class SafeEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, float) and np.isnan(obj):
            return None
        if isinstance(obj, (np.integer,)):
            return int(obj)
        if isinstance(obj, (np.floating,)):
            return float(obj)
        if isinstance(obj, (np.bool_,)):
            return bool(obj)
        return super().default(obj)

OUTPUT_PATH = r"C:\lecture\flow\00_data\02_processed\ingredient_merged2.json"

records = df_merged.where(pd.notna(df_merged), None).to_dict(orient="records")

with open(OUTPUT_PATH, "w", encoding="utf-8") as f:
    json.dump(records, f, ensure_ascii=False, indent=2, cls=SafeEncoder)

print(f"저장 완료: {OUTPUT_PATH}")
print(f"총 행 수: {len(records)}")
print(f"\n[샘플 1개]")
print(json.dumps(records[0], ensure_ascii=False, indent=2, cls=SafeEncoder))
```
