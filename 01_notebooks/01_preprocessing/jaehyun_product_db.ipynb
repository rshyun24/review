```python
# ── product_db: 화해 원본 CSV에서 직접 생성 ──────────────
df_product = pd.read_csv(
    r"C:\lecture\flow\00_data\00_raw\hwahae_all.csv",
    encoding="utf-8"
)

product_cols = [
    "korean",
    "product_id",
    "product_name",
    "brand_name",
    "ingredient_count",
    "primary_attr",
    "price",
    "consumer_price",
    "avg_ratings",
    "review_count",
    "topics_positive",
    "topics_negative",
]

df_product = df_product[[c for c in product_cols if c in df_product.columns]]

# 컬럼 리네임
df_product = df_product.rename(columns={
    "korean":            "ingredient_ko",
    "product_id":        "hw_product_id",
    "product_name":      "hw_product_name",
    "brand_name":        "hw_brand_name",
    "ingredient_count":  "hw_ingredient_count",
    "primary_attr":      "hw_primary_attr",
    "price":             "hw_price",
    "consumer_price":    "hw_consumer_price",
    "avg_ratings":       "hw_avg_ratings",
    "review_count":      "hw_review_count",
    "topics_positive":   "hw_topics_positive",
    "topics_negative":   "hw_topics_negative",
})

df_product.to_csv(
    r"C:\lecture\flow\00_data\02_processed\product_db.csv",
    index=False,
    encoding="utf-8-sig"
)
print(f"product_db 저장 완료: {len(df_product)}행")
```
