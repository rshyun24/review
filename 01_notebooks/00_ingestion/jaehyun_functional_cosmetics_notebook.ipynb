```python
import requests
import pandas as pd

API_KEY = "SKIN_API"
END_POINT = "https://apis.data.go.kr/1471000/FtnltCosmRptPrdlstInfoService/getRptPrdlstInq"

all_items = []
page = 1

while True:
    url = f"{END_POINT}?serviceKey={API_KEY}&type=json&numOfRows=100&pageNo={page}"

    response = requests.get(url)
    print(f"페이지 {page} - 상태코드: {response.status_code}")

    # 응답 확인
    if response.status_code != 200:
        print("오류 응답:", response.text[:300])
        break

    data = response.json()

    # 첫 페이지에서 구조 확인
    if page == 1:
        print("응답 키:", list(data.keys()))

    body = data.get('body', {})
    items = body.get('items', [])

    if not items:
        print("데이터 없음. 종료.")
        break

    all_items.extend(items)

    # 전체 데이터 수 확인 후 마지막 페이지 판단
    total_count = body.get('totalCount', 0)
    print(f"  → {len(items)}개 수집 (누적: {len(all_items)} / 전체: {total_count})")

    if len(all_items) >= total_count:
        print("전체 수집 완료!")
        break

    page += 1

# CSV 저장
df = pd.DataFrame(all_items)
df.to_csv("기능성화장품_보고품목정보.csv", index=False, encoding="utf-8-sig")
print(f"\n✅ 저장 완료! 총 {len(df)}행 × {len(df.columns)}열")
print("저장 위치: 기능성화장품_보고품목정보.csv")
```
